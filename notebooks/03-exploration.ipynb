# Exploration des données éco2mix


Objectif : avoir une vue d'ensemble sur la consommation française d'électricité


```python
import pandas as pd

df = pd.read_csv("../data/eco2mix_2023_2024.csv", sep=";")
```

```python
df = df.dropna(subset=["consommation"])
```

```python
df["date_heure"] = pd.to_datetime(df["date_heure"])
```

```python
df['heure']=df["date_heure"].dt.hour
```

```python
df['jour_semaine']=df['date_heure'].dt.dayofweek
```

```python
df['mois']=df['date_heure'].dt.month
```

```python
df.head()
```

## Vue d'ensemble sur les deux ans


```python
df.plot(x='date_heure',y='consommation',figsize=(15,5))
```

On remarque trois pics pendant cette période du début 2023 jusqu'au début 2025. Le premier début 2023, le deuxième fin 2023/début 2024, et le troisième fin 2024/début 2025 : ce sont les trois hivers que notre fenêtre de deux ans attrape.

## Le profil d'une journée moyenne

```python
df.groupby('heure')['consommation'].mean().plot(title='Jour moyen',kind='bar')
```

Le premier pic de consommation journalière est vers 11h, le deuxième est vers 18h (en horaire UTC, soit environ 12h-13h et 19h-20h en heure française), c'est la pointe du déjeuner et la pointe du soir.
La nuit reste la tranche de temps avec le moins de consommation.

##  L'effet du jour de la semaine

```python
df.groupby('jour_semaine')['consommation'].mean().plot(kind='bar',title='Semaine moyenne')
```

```python
df.groupby('jour_semaine')['consommation'].mean()
```

```python
df.groupby('jour_semaine')['consommation'].mean().max()/df.groupby('jour_semaine')['consommation'].mean().min()
```

```python
1-df.groupby('jour_semaine')['consommation'].mean()[0]/df.groupby('jour_semaine')['consommation'].mean()[0:4].mean()
```

```python
1-df.groupby('jour_semaine')['consommation'].mean()[5:6].mean()/df.groupby('jour_semaine')['consommation'].mean()[0:4].mean()
```

Globalement, tous les jours de la semaine sont assez proches en termes de consommation, avec le dimanche comme jour le plus faible (45 383 MW) et le mardi comme jour le plus fort (51 901 MW), soit 14% de plus que le dimanche. On remarque aussi qu'en semaine, le lundi est le jour avec le moins de consommation (50 669 MW), soit 1,6% de moins que la moyenne des jours ouvrés, cela peut être justifié par le fait que beaucoup de jours fériés tombent un lundi, ou encore que l'industrie redémarre ce jour-là progressivement et non pas brusquement. Le weekend reste la tranche de la semaine avec le moins de consommation (8,6% de moins que la moyenne en semaine), ce qui peut être expliqué par l'inactivité d'une grosse partie de l'industrie.

## Les saisons

```python
df.groupby('mois')['consommation'].mean().plot(kind='bar',title='Année moyenne par mois')
```

```python
df.groupby('mois')['consommation'].mean()
```

```python
1-df.groupby('mois')['consommation'].mean().min()/df.groupby('mois')['consommation'].mean().max()
```

Le mois avec le plus de consommation est janvier (63 995 MW) et celui avec le moins est août (41 420 MW), soit 35% de moins que janvier. On remarque alors que pendant la période printemps-été il y a une baisse nette de consommation par rapport à la période automne-hiver, cela laisse à penser que cette variation est engendrée par l'utilisation du chauffage pendant les périodes froides
